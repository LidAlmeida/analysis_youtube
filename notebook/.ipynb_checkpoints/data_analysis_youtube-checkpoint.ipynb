# FUNÇÕES

```python
def get_video_ids(youtube, playlist_id):
    """
    Obtém uma lista de IDs de vídeo de todos os vídeos em uma determinada lista de reprodução

    :param youtube: objeto de construção de googleapiclient.discovery
    :param playlist_id: ID da lista de reprodução do canal
    :return: Lista de IDs de vídeo de todos os vídeos na lista de reprodução
    """

    request = youtube.playlistItems().list(
        part='contentDetails',
        playlistId=playlist_id,
        maxResults=50
    )
    response = request.execute()

    video_ids = []

    for item in response['items']:
        video_ids.append(item['contentDetails']['videoId'])

    next_page_token = response.get('nextPageToken')
    more_pages = True

    while more_pages:
        if next_page_token is None:
            more_pages = False
        else:
            request = youtube.playlistItems().list(
                part='contentDetails',
                playlistId=playlist_id,
                maxResults=50,
                pageToken=next_page_token
            )
            response = request.execute()

            for item in response['items']:
                video_ids.append(item['contentDetails']['videoId'])

            next_page_token = response.get('nextPageToken')

    return video_ids
```

```python
def get_video_details(youtube, video_ids):
    """
    Obtém as estatísticas de todos os vídeos com IDs fornecidos.
    Parâmetros:

    youtube: objeto build da biblioteca googleapiclient.discovery
    video_ids: lista de IDs de vídeo

    Retorna:
    Um dataframe com as estatísticas dos vídeos, ou seja:
        'channelTitle', 'title', 'description', 'tags', 'publishedAt'
        'viewCount', 'likeCount', 'favoriteCount', 'commentCount'
        'duration', 'definition', 'caption'
    """

    all_video_info = []
    
    # A cada 50 IDs de vídeo, uma nova solicitação é feita para obter as informações do vídeo
    for i in range(0, len(video_ids), 50):
        request = youtube.videos().list(
            part="snippet,contentDetails,statistics",
            id=','.join(video_ids[i:i+50])
        )
        response = request.execute() 

        # Para cada vídeo, o dicionário "stats_to_keep" é utilizado para manter apenas as informações relevantes
        for video in response['items']:
            stats_to_keep = {'snippet': ['channelTitle', 'title', 'description', 'tags', 'publishedAt'],
                             'statistics': ['viewCount', 'likeCount', 'favoriteCount', 'commentCount'],
                             'contentDetails': ['duration', 'definition', 'caption']
                            }
            video_info = {}
            video_info['video_id'] = video['id']

            # Para cada estatística relevante, tenta-se extrair do dicionário de informações do vídeo,
            # caso contrário, a estatística é armazenada como None
            for k in stats_to_keep.keys():
                for v in stats_to_keep[k]:
                    try:
                        video_info[v] = video[k][v]
                    except:
                        video_info[v] = None

            all_video_info.append(video_info)

    # Todas as informações coletadas são armazenadas em um dataframe e retornadas
    return pd.DataFrame(all_video_info)
```

```python
def get_comments_in_videos(youtube, video_ids):
    """
    Obter comentários de primeiro nível em forma de texto de todos os vídeos com os IDs fornecidos (apenas os primeiros 10 comentários devido ao limite de cota da API do YouTube)

    Parâmetros:
    youtube: objeto build da googleapiclient.discovery
    video_ids: lista de IDs de vídeo

    Retorna:
    Dataframe com os IDs dos vídeos e o comentário de primeiro nível associado em texto.
    """
    all_comments = []

    for video_id in video_ids:
        try:
            # Requisição para obter os comentários do vídeo
            request = youtube.commentThreads().list(
                part="snippet,replies",
                videoId=video_id
            )
            response = request.execute()

            # Extrair os comentários de primeiro nível e convertê-los em texto
            comments_in_video = [comment['snippet']['topLevelComment']
                                 ['snippet']['textOriginal'] for comment in response['items'][0:10]]

            # Armazenar as informações do vídeo e os comentários em um dicionário
            comments_in_video_info = {
                'video_id': video_id, 'comments': comments_in_video}
            all_comments.append(comments_in_video_info)

        except:
            # Quando ocorre um erro - provavelmente porque os comentários estão desativados em um vídeo
            print('Não foi possível obter os comentários para o vídeo ' + video_id)

    # Converter a lista de dicionários em um DataFrame
    return pd.DataFrame(all_comments)
```

```python
def get_comments_in_videos(youtube, video_ids):
    """
    Obter comentários de primeiro nível em forma de texto de todos os vídeos com os IDs fornecidos (apenas os primeiros 10 comentários devido ao limite de cota da API do YouTube)

    Parâmetros:
    youtube: objeto build da googleapiclient.discovery
    video_ids: lista de IDs de vídeo

    Retorna:
    Dataframe com os IDs dos vídeos e o comentário de primeiro nível associado em texto.
    """
    all_comments = []

    for video_id in video_ids:
        try:
            # Requisição para obter os comentários do vídeo
            request = youtube.commentThreads().list(
                part="snippet,replies",
                videoId=video_id
            )
            response = request.execute()

            # Extrair os comentários de primeiro nível e convertê-los em texto
            comments_in_video = [{'comment_text': comment['snippet']['topLevelComment']['snippet']['textOriginal'],
                                  'likes': comment['snippet']['topLevelComment']['snippet']['likeCount']}
                                 for comment in response['items'][0:10]]

            # Armazenar as informações do vídeo e os comentários em um dicionário
            comments_in_video_info = {
                'video_id': video_id, 'comments': comments_in_video}
            all_comments.append(comments_in_video_info)

        except:
            # Quando ocorre um erro - provavelmente porque os comentários estão desativados em um vídeo
            print('Não foi possível obter os comentários para o vídeo ' + video_id)

    # Converter a lista de dicionários em um DataFrame
    return pd.DataFrame(all_comments)
```

Ao analisar a wordcloud, é possível notar que certas palavras têm um tamanho maior em comparação com outras, o que sugere que essas palavras foram mencionadas com mais frequência nos títulos analisados. Termos como data, dado, python, estatística e cientista se destacam na nuvem de palavras, indicando que foram mencionados com maior frequência.

Por outro lado, algumas palavras aparecem em tamanho menor, o que sugere que foram mencionadas com menos frequência nos títulos. Palavras como algoritmo, classificação e regressão são exemplos de termos que aparecem em menor tamanho na wordcloud. Esses resultados podem ser úteis para entender quais são os assuntos mais abordados nos títulos e quais são menos frequentes.

```python
# Concatenando todos os títulos em uma única string
text = ' '.join(df_video['title'].astype(str))

# Removendo os caracteres especiais e números
text = re.sub('[^a-zA-ZÀ-ú]', ' ', text)

# Convertendo para letras minúsculas
text = text.lower()

# Removendo as stopwords
stopwords_pt = set(nltk.corpus.stopwords.words('portuguese'))
stopwords_en = set(nltk.corpus.stopwords.words('english'))
stopwords = stopwords_pt.union(stopwords_en)

words = [word for word in text.split() if word not in stopwords]

# Criando a WordCloud
wordcloud = WordCloud(width=2000, height=1000, background_color='black', random_state=1,
                      stopwords=stopwords, collocations=False, colormap='YlGnBu').generate(text)

# Plotando a WordCloud
plt.figure(figsize=(20, 16))
plt.imshow(wordcloud, interpolation='bilinear')
plt.axis('off')
```
